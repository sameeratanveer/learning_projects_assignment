### --------------- 3. Uni-variate Analysis : ---------------------

#### 1. Numerical Features : 

###### 01_ : Age : 

```python
plt.figure(figsize=(9,5))
plt.hist(df['Age'], bins=30,edgecolor='black', color='g', alpha=0.4)
plt.xticks(np.arange(23,55,1), rotation=90)
plt.yticks(np.arange(1,25,1))
plt.xlabel("Age of the Employee")
plt.ylabel("Count of Age")
plt.title("Age vs count of the Employees")
plt.show()
```

###### 02_ : Years of experience 

```python
plt.figure(figsize=(9,7))
plt.hist(df['Years of Experience'], bins=25,edgecolor='black', color='y', alpha=0.4)
plt.xticks(np.arange(0,25,1), rotation=90)
plt.yticks(np.arange(0,30,1))
plt.xlabel("Years of Experience of the Employee")
plt.ylabel("Count")
plt.title("Years of Experience vs count of the Employees")
plt.show()
```

```python
sns.kdeplot(data=df, x='Years of Experience')
```

###### 03_ : Salary 

```python
plt.figure(figsize=(9,7))
plt.hist(df['Salary'], bins=30,edgecolor='black', color='b', alpha=0.4)
plt.xticks(np.arange(0,260000,10000), rotation=90)
plt.yticks(np.arange(0,45,2))
plt.xlabel("Salary of the Employee")
plt.ylabel("Count")
plt.title("Salary vs count of the Employees")
plt.show()
```

#### 2. Categorical Variables :
